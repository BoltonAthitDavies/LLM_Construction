## 1. tokenization

```python
# Read JSON file
df = pd.DataFrame()
for list in json_list:
    print(list)
    with open(os.path.join(json_path, list), 'r', encoding='utf-8') as file:
        data = json.load(file)
        des = data[0]['description'].split("\n")   
        # print(des)
    tokens = []
    for d in des:
        token = word_tokenize(d, engine="newmm", keep_whitespace=True)
        tokens.extend(token)
    # print(tokens)
    data_char = pd.DataFrame({'doc name': json_path.split("\\")[-1],
                              'page' : list[5:7].replace(".",""),
                              'content' : "",
                              'text': data[0]['description'],
                              'tokens': [tokens],
                              'tokens_count': len(tokens)})
    df = pd.concat([df, data_char], ignore_index=True)
```

### 2.2 Unique word

```python
tags = []
for words in df['tokens']:
    print(words)
    tag = pos_tag(words, corpus="orchid", engine="perceptron")
    tags.append(tag)

df['pos_tags'] = tags
```

### 2.1 Language

```python
# from deep_translator import GoogleTranslator
# th2en = GoogleTranslator(source="th", target="en")

# th2en.translate("มหาชน")
```

```python
from deep_translator import GoogleTranslator
th2en = GoogleTranslator(source="th", target="en")
langs = []
th_cnts = []
en_cnts = []
for words in df['tokens']:
    print(words)
    th_cnt = 0
    en_cnt = 0
    lang_bag = []
    for token in words:
        print(token)
        lang = th2en.translate(token)
        if lang == token:
            lang = "eng"
            en_cnt += 1
        else:
            lang = "th"
            th_cnt += 1
        lang_bag.append((token, lang))
    langs.append(lang_bag)
    th_cnts.append(th_cnt)
    en_cnts.append(en_cnt)

df['translated_tokens'] = langs
df['th_token_count'] = th_cnts
df['en_token_count'] = en_cnts
```

```python
from FlagEmbedding import BGEM3FlagModel
model = BGEM3FlagModel('BAAI/bge-m3',  
                       use_fp16=True) # Setting use_fp16 to True speeds up computation with a slight performance degradation
embedded_vectors = []
for t in df['tokens']:
    text = ''
    for token in t:
        text += token
    print(text)
    embedded_vector = model.encode([text])['dense_vecs']
    embedded_vectors.append(embedded_vector)
df['embedded_vectors'] = embedded_vectors
```
